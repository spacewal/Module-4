## 1.
### Read the data from the CSV file into a DataFrame and display the first five rows.

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/fenago/dw/main/ramen-ratings.csv")
```

```python
df.head(5)
```

## 2.
### Add a column to the DataFrame that stores the average rating for each country. Then, display the first five rows and verify that the column was added and that its values appear to be correct.

```python
df["avg_rating"] = df.groupby("Country")["Stars"].transform(func="mean")

df[["Country", "Stars", "avg_rating"]].head(5)
```

## 3.
### Group the data by the Country and Style columns and take the mean. Then, unstack the Style column and display the data for the Stars column.

```python
df.groupby(['Country', 'Style']).mean().unstack(level='Style')['Stars']
```

## 4.
### Write a function that takes a row as input and returns a shortened version of the values in the Variety column. The values should consist of the first two words of the variety name, followed by a space and an ellipsis (…). Hint: You’ll need to split the values into a list of words, get the first two words in the list, add the ellipsis to the list, and then join the words in the list.

```python
df.head(5)
```

```python
def shorten_variety(row):
    words = row['Variety'].split()  # Split the value into a list of words
    shortened = words[:2]  # Get the first two words
    shortened.append('…')  # Add the ellipsis
    return ' '.join(shortened)  # Join the words into a string
```

## 5.
### Apply the function you just wrote to the DataFrame, and store the results in a column named ShortVariety. Then, display the first five rows to be sure this worked.

```python
df['ShortVariety'] = df.apply(shorten_variety, axis=1)
```

```python
df.head(5)
```
