# Data Wrangling

```python
df.head()
```

```python
# Remove rows where any column contains 'Mega'
df = df[~df.apply(lambda row: row.astype(str).str.contains('Mega').any(), axis=1)]
```

```python
df.head()
```

```python
df.info()
```

```python
# Create a new column 'Average' that is the mean of the specified columns
df['Average'] = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].mean(axis=1)
```

```python
df.head()
```

```python
df.sort_values(by='Average')
```

## Calculating Total Stats

```python
# Lambda to calculate total stats
total_stats = lambda row: row['HP'] + row['Attack'] + row['Defense'] + row['Sp. Atk'] + row['Sp. Def'] + row['Speed']
```

```python
df['Total Stats'] = df.apply(total_stats, axis=1)
```

## Classifying PokÃ©mon by Stat Threshold

```python
# Lambda for classifying PokÃ©mon based on their Attack value
classify_by_attack = lambda attack: 'Weak' if attack <= 50 else ('Strong' if attack <= 100 else 'Very Strong')
```

```python
df['Classification'] = df['Attack'].apply(classify_by_attack)
```

```python
df.head()
```

```python
df.tail()
```

## Top 5 Strongest and Top 5 Weakest PokÃ©mon

```python
# Finding the top 5 strongest PokÃ©mon
top_5_strongest = df.sort_values(by='Total Stats', ascending=False).head(5)

# Finding the top 5 weakest PokÃ©mon
top_5_weakest = df.sort_values(by='Total Stats', ascending=True).head(5)

# Display
print("Top 5 Strongest PokÃ©mon:")
print(top_5_strongest)

print("\nTop 5 Weakest PokÃ©mon:")
print(top_5_weakest)
```

```python
# Group by 'Type 1' and 'Type 2' and calculate the mean for each group
average_stats_by_type = df.groupby(['Type 1', 'Type 2']).mean().reset_index()

# Rename the column to distinguish it from the original 'Total Stats'
average_stats_by_type = average_stats_by_type.rename(columns={'Total Stats': 'Average Total Stats'})

# Merge the average stats back onto the original DataFrame based on 'Type 1' and 'Type 2'
df = pd.merge(df, average_stats_by_type[['Type 1', 'Type 2', 'Average Total Stats']], on=['Type 1', 'Type 2'], how='left')

df.head()
```
